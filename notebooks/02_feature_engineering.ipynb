```python
import sys
import os
import pandas as pd

# Thiết lập đường dẫn để import từ thư mục src
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src.data.cleaner import HotelCleaner
from src.mining.association import discretize_lead_time # Hàm này bạn cần định nghĩa trong file association.py
```

```python
# Load dữ liệu gốc
raw_data = pd.read_csv("../data/raw/hotel_bookings.csv")

# Làm sạch dữ liệu (loại bỏ leakage)
cleaner = HotelCleaner(raw_data)
df_cleaned = cleaner.clean_data()

print("Kích thước dữ liệu sau khi làm sạch:", df_cleaned.shape)
```

```python
# Gọi hàm rời rạc hóa đã import thành công ở Cell 1
df_prepared = discretize_lead_time(df_cleaned)

# Kiểm tra kết quả
print("Các cột hiện có:", df_prepared.columns.tolist())
print("5 dòng đầu sau khi rời rạc hóa lead_time:")
print(df_prepared[['lead_time', 'lead_time_cat']].head())
```
